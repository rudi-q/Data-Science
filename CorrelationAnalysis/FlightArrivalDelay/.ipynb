<h2>AIRLINE DATASET</h2>

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import time
```

```python
start=time.time()
data=pd.read_csv(
    "airlinedataset2.csv",
    usecols= ["CRS_DEP_TIME", "DEP_TIME", "DEP_DELAY", "TAXI_OUT", "WHEELS_OFF",
              "WHEELS_ON", "TAXI_IN", "CRS_ARR_TIME", "ARR_TIME", "ARR_DELAY",
              "CANCELLED", "DIVERTED", "CRS_ELAPSED_TIME",
              "ACTUAL_ELAPSED_TIME", "AIR_TIME", "DISTANCE", "CARRIER_DELAY",
              "WEATHER_DELAY", "NAS_DELAY", "SECURITY_DELAY", "LATE_AIRCRAFT_DELAY"]
)
end=time.time()
print("This took ",end-start," seconds")
print("Datashape",data.shape)
```

```python
#x=data["Missing Count"].median
data=data.dropna()
data
```

```python
correlation_matrix = data.corr()
# Select the correlation of arrival_delay with other attributes
arrival_delay_correlation = correlation_matrix['ARR_DELAY']

# Drop the arrival_delay column since we're interested in correlations with it
arrival_delay_correlation = arrival_delay_correlation.drop('ARR_DELAY')

# Sort the correlations in descending order
arrival_delay_correlation = arrival_delay_correlation.sort_values(ascending=False)

# Plot the correlation using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(arrival_delay_correlation.to_frame(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation with Arrival Delay')
plt.xlabel('Attribute')
plt.ylabel('Arrival Delay')
plt.show()
```

```python
# Sample data
x = data["DEP_DELAY"]
y = data["ARR_DELAY"]

# Create scatter plot
plt.scatter(x, y)

# Set labels and title
plt.xlabel('Departure Delay')
plt.ylabel('Arrival Delay')
plt.title('Scatter Plot')

# Display the plot
plt.show()
```
